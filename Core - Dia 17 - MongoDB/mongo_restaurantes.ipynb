```python
# Devolver ID del restaurante, nombre, distrito y la cocina para aquellos 
# restaurantes que pertenecen al distrito de Staten Island o Queens o Bronx o Brooklyn

projection = {"_id":1, "name": 1,  "borough": 1, "cuisine":1} # COLUMNAS (SELECT)
filter = {"borough": "Bronx", "borough": "Brooklyn", "borough": "Queens", "borough": "Staten Island"} # FILAS (WHERE)

list(restaurantes.find(filter=filter, projection=projection).limit(20))
```

```python
# Devolver ID del restaurante, nombre, distrito y la cocina de aquellos restaurantes 
# que lograron una puntuación que no supere los 10

projection = {"_id":1, "name": 1,  "borough": 1, "cuisine":1} # COLUMNAS (SELECT)
filter = { "grades.score": {"$lt":10}} # FILAS (WHERE)

list(restaurantes.find(filter=filter, projection=projection).limit(20))
```

```python
# Devolver ID del restaurante, el nombre y las calificaciones del restaurante para aquellos
# restaurantes que obtuvieron una calificación de "A" y obtuvieron un puntaje de 11 en una 
# fecha ISO "2014-08-11T00: 00: 00Z" entre muchas fechas de encuesta

projection = {"_id":1, "name": 1,  "grades.score": 1} # COLUMNAS (SELECT)
filter = { "grades.score": 11,  "grades.grade": "A","grades.date":1407708000 } # FILAS (WHERE)

list(restaurantes.find(filter=filter, projection=projection).limit(20))
```

```python
# Proyección: especifica qué campos queremos incluir en los resultados
projection = {
    "_id": 1, 
    "name": 1,  
    "grades": 1  # Aquí opté por incluir todo el subdocumento 'grades' para claridad
}

# Filtro: usando $elemMatch para buscar documentos que tengan al menos un elemento en 'grades' que coincida con las condiciones
filter = {
    "grades": {
        "$elemMatch": {
            "score": 11, 
            "grade": "A", 
            "date": 1407708000
        }
    }
}

# Ejecutar la consulta
list(restaurantes.find(filter=filter, projection=projection))
```
