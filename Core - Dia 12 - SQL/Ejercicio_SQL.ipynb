```python
# Conectamos con la base de datos chinook.db
connection = sqlite3.connect("data/chinook.db")
#SI AL LLAMAR AL ARCHIVO NO ESTAMOS DONDE TOCA, LO VA A CREAR

# Obtenemos un cursor que utilizaremos para hacer las queries
crsr = connection.cursor() #LA CONEXION ES EL PUENTE Y EL CURSOS ES EL QUE VA Y TRAE LOS DATOS
```

```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    crsr.execute(query)

    # Almacena los datos de la query 
    ans = crsr.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in crsr.description]

    return pd.DataFrame(ans,columns=names)
```

```python
# 6. Muestra todas las conciones con su nombre, formato, album y genero 
query = '''
SELECT tracks.Name as Track_Name, media_types.Name as Media_Format, albums.Title as Album_Title,
genres.Name as Genre

FROM tracks
JOIN media_types
  ON tracks.MediaTypeId = media_types.MediaTypeId
JOIN albums
  ON tracks.AlbumId = albums.AlbumId
JOIN genres
  ON tracks.GenreId = genres.GenreId

ORDER BY albums.Title ASC
'''

sql_query(query)
```

```python
# 6. Muestra todas las conciones con su nombre, formato, album y genero 
query = '''
SELECT 
    t.Name AS SongName,
    mt.Name AS Format,
    a.Title AS Album,
    g.Name AS Genre
FROM 
    tracks t
JOIN 
    media_types mt
ON 
    t.MediaTypeId = mt.MediaTypeId
JOIN 
    albums a
ON 
    t.AlbumId = a.AlbumId
JOIN 
    genres g
ON 
    t.GenreId = g.GenreId;
'''
sql_query(query)
```

```python
# 7. Muestra cuantas canciones hay en cada playlist y el nombre de cada playlist 

query = '''
SELECT 
    p.Name AS PlaylistName,
    COUNT(pt.TrackId) AS TotalSongs
FROM 
    playlists p
JOIN 
    playlist_track pt
ON 
    p.PlaylistId = pt.PlaylistId
GROUP BY 
    p.Name;
'''

sql_query(query)
```

```python
# 8. ¿Cuánto ha vendido cada empleado?
query = '''
SELECT 
    e.EmployeeId,
    e.FirstName || ' ' || e.LastName AS EmployeeName,
    SUM(ii.UnitPrice * ii.Quantity) AS TotalSales
FROM 
    invoices i
JOIN 
    invoice_items ii
ON 
    i.InvoiceId = ii.InvoiceId
JOIN 
    employees e
ON 
    invoices.SupportRepId = e.EmployeeId
GROUP BY 
    e.EmployeeId, e.FirstName, e.LastName
ORDER BY 
    TotalSales DESC;
    '''
sql_query(query)
```

```python
# 9. ¿Quién ha sido el agente de ventas que más ha vendido en 2009?
query = '''
SELECT 
    e.EmployeeId,
    e.FirstName || ' ' || e.LastName AS EmployeeName,
    SUM(ii.UnitPrice * ii.Quantity) AS TotalSales
FROM 
    invoices i
JOIN 
    invoice_items ii
ON 
    i.InvoiceId = ii.InvoiceId
JOIN 
    employees e
ON 
    i.SupportRepId = e.EmployeeId
WHERE 
    strftime('%Y', i.InvoiceDate) = '2009'
GROUP BY 
    e.EmployeeId, e.FirstName, e.LastName
ORDER BY 
    TotalSales DESC
LIMIT 1;

    '''
sql_query(query)
```

```python
 # 10. ¿Quiénes son los 3 grupos que más han vendido?

query = '''
SELECT 
    a.ArtistId,
    a.Name AS ArtistName,
    SUM(ii.UnitPrice * ii.Quantity) AS TotalSales
FROM 
    tracks t
JOIN 
    albums al
ON 
    t.AlbumId = al.AlbumId
JOIN 
    artists a
ON 
    al.ArtistId = a.ArtistId
JOIN 
    invoice_items ii
ON 
    t.TrackId = ii.TrackId
GROUP BY 
    a.ArtistId, a.Name
ORDER BY 
    TotalSales DESC
LIMIT 3;

    '''
sql_query(query)
```
