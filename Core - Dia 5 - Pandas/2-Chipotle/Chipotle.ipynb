### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

```python
chip = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep = "\t")
chip
   
```

### Step 3. Name of the max valued product

```python
chip.info()
```

```python
chip.sort_values("item_price")
```

```python
chip["item_price"] = chip["item_price"].str[1:].astype(float)
chip
```

```python
chip_max = chip.sort_values(by= "item_price", ascending= False)

chip_max[0:1]
```

### Step 4. How many products cost more than $10.00?

```python
mask = chip['item_price'] > 10
mask
```

```python
chip_mayor10 = chip.loc[mask,:]

print(f"De un total de {chip.shape[0]}, {chip_mayor10.shape[0]} son mayores a 10$.")
chip_mayor10
```

Extra: CuÃ¡ntos pedidos se han hecho con un producto de mÃ¡s de 10$

```python
mask_ud = chip["quantity"] == 1
chip_ud = chip.loc[mask_ud,:]
chip_ud = chip_ud.loc[mask,:]
print(f"Se han realizado un total de {chip_ud.shape[0]} pedidos de 1 ud. con valor mayor a 10$.")
chip_ud
```

### Step 5. What is the price of each item? Get only item_name and unit_price

```python
# Crea una nueva columna llamada unit_price
chip.loc[:,"unite_price"] = chip.loc[:,"item_price"] / chip.loc[:,"quantity"]
chip
```

### Step 6. Sort by the name of the item

```python
chip = chip.sort_values(by= "item_name")
chip
```

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
chip.sort_values(by= "unite_price", ascending=False).iloc[0]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
mask_salad = chip["item_name"] == "Veggie Salad Bowl"

chip_salad = chip.loc[mask_salad,:]
print(f"La Veggie Salad Bowl se ha pedido un total de {chip_salad.shape[0]} veces")
chip_salad
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
mask_soda = chip["item_name"] == "Canned Soda"
chip_soda = chip.loc[mask_soda,:]
chip_soda
```

```python
mask_soda2 = chip_soda["quantity"] >1
mask_soda2
```

```python
chip_soda_morethanone = chip_soda.loc[mask_soda2,:]
print(f"Se han realizado un total de {chip_soda_morethanone.shape[0]} pedidos con mas de una Canned Soda.")
chip_soda_morethanone
```
