```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users 
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation")[["age"]].mean().sort_values("age")
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# Debemos divitir el numero de Males de cada oficio entre el total de users de cada oficio.
users
```

```python
#Creamos un DF con el total de usuarios por ocupacion
total_users_ocup = users.groupby("occupation")[["gender"]].count()
total_users_ocup = total_users_ocup.reset_index()
total_users_ocup = total_users_ocup.rename(columns={"gender" : "total"})
total_users_ocup
```

```python
#sacamos el numero de males por ocupacion
mask_males = users.loc[:,"gender"] == "M"
male_occupations = users.loc[mask_males,:]

mask_fem = users.loc[:,"gender"] == "F"
fem_occupations = users.loc[mask_fem,:]
fem_occupations = fem_occupations.groupby("occupation")[["gender"]].count().rename(columns={"gender":"females"}).reset_index()

male_occupations = male_occupations.groupby("occupation")[["gender"]].count().rename(columns={"gender":"males"}).reset_index()
male_occupations
```

```python
#añadimos el total dambos sexos al dataframe anterior y creamos una nueva columna con la diviison de ambos
user_males = pd.merge(left=male_occupations, right=total_users_ocup)
user_males.loc[:,"percentage_males"] =  user_males.loc[:,"males"] / user_males.loc[:,"total"]
user_males
```

```python
users_total = pd.merge(left=user_males, right=fem_occupations, how='outer')
users_total.loc[:,'females'] = users_total.loc[:,'females'].fillna(0).astype(int)

users_total.loc[:,"percentage_females"] = users_total.loc[:,"females"] / users_total.loc[:,"total"]
users_total = users_total.iloc[:,[0,2,1,3,4,5]]
users_total
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users_ages_min = users.groupby("occupation")["age"].min().reset_index().rename(columns={"age":"age_min"})
users_ages_max = users.groupby("occupation")["age"].max().reset_index().rename(columns={"age":"age_max"})
users_ages = pd.merge(left=users_ages_min, right=users_ages_max, how="outer")
users_ages
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users_mean_age = users.groupby(["occupation", "gender"])["age"].mean().reset_index()

users_mean_age
```

### Step 8.  For each occupation present the percentage of women and men

```python
users_ocupp_gender = users.groupby("occupation")["gender"].value_counts(normalize=True).reset_index()
users_ocupp_gender
```
